```python
import pandas as pd
from sklearn import tree
```

```python
dataset= pd.read_csv("C:/Users/ASUS/general_data.csv")
```

```python
dataset.head()
```

```python
from sklearn import preprocessing 
```

```python
lr = preprocessing.LabelEncoder() 
```

```python
dataset['EducationField']= lr.fit_transform(dataset['EducationField']) 
dataset['Attrition']= lr.fit_transform(dataset['Attrition']) 
dataset['BusinessTravel']= lr.fit_transform(dataset['BusinessTravel']) 
dataset['Department']= lr.fit_transform(dataset['Department']) 
dataset['Gender']= lr.fit_transform(dataset['Gender']) 
dataset['JobRole']= lr.fit_transform(dataset['JobRole']) 
dataset['MaritalStatus']= lr.fit_transform(dataset['MaritalStatus']) 
dataset['Over18']= lr.fit_transform(dataset['Over18']) 
```

```python
dataset.head()
```

#### Deleted the columns not so important 

```python
dataset1=dataset.drop(['EmployeeID'], axis = 1) 
dataset1
```

#### Checked for Null Values

```python
dataset1.isnull().sum()
```

#### Handling the missing values

```python
dataset1=dataset1.dropna(axis=0)
dataset1.isnull().sum()
```

#### mapped the important columns to be worked on to a variable

```python
predictor=pd.DataFrame([dataset1['Age'],dataset1['BusinessTravel'],dataset1['Department'],dataset1['DistanceFromHome'],dataset1['Education'],dataset1['EducationField'],dataset1['EmployeeCount'],dataset1['Gender'],dataset1['JobLevel'],dataset1['JobRole'],dataset1['MaritalStatus'],dataset1['MonthlyIncome'],dataset1['NumCompaniesWorked'],dataset1['Over18'],dataset1['PercentSalaryHike'], dataset1['StandardHours'], dataset1['StockOptionLevel'],dataset1['TotalWorkingYears'], dataset1['TrainingTimesLastYear'], dataset1['YearsAtCompany'],dataset1['YearsSinceLastPromotion'], dataset1['YearsWithCurrManager']]).T
```

#### Now to select the most useful colums for classification, we use randomforestclassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
features=['Age','BusinessTravel','Department','DistanceFromHome','Education','EducationField','EmployeeCount','Gender','JobLevel','Education','JobRole','MaritalStatus','MonthlyIncome','NumCompaniesWorked','Over18','PercentSalaryHike','StandardHours','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager']
```

```python
rf_model.fit(X=predictor,y=dataset1['Attrition'])
```

```python
print('The oob score is\n',rf_model.oob_score_)
```

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(feature,imp)
```

### According to this data we selected the features and mapped then to predictor

```python
predictor=pd.DataFrame([dataset1['Age'], dataset1['DistanceFromHome'], dataset1['MonthlyIncome'], dataset1['PercentSalaryHike'], dataset1['YearsAtCompany']]).T
```

```python
model=tree.DecisionTreeClassifier(max_depth=5)
```

```python
model.fit(X=predictor,y=dataset1['Attrition'])
```

```python
with open("AttritionDt.dot",'w')as f:
    f= tree.export_graphviz(model, feature_names=['DistanceFromHome','Age','MonthlyIncome','PercentSalaryHike','YearsAtCompany'],out_file=f);
```

```python
model.score(X=predictor,y=dataset1['Attrition'])
```
