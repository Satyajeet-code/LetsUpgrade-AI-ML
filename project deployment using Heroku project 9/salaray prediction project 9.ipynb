```python

# import packages

import warnings
warnings.simplefilter('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
# pickle is used to save the model created by us
```

```python
df = pd.read_csv("../input/salary-prediction/hiring.csv")
df.head()
```

```python
df.isna().sum()
```

```python
# experience
df['experience'].fillna(0,inplace = True)
df.isna().sum()
```

```python
df['test_score'].fillna(df['test_score'].mean(),inplace = True)
df.isna().sum()
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]  #all the columns except -1th column as dependent variables
X.head()
```

```python
X.shape
```

```python
X.experience
```

```python
# convert text into intergers
def conv(x):
    dict = {'two':2,
            'three':3,
            'five':5,
            'seven':7,
            'ten':10,
            'eleven':11,
            0:0 }
    return dict[x]
```

```python
X['experience'] = X['experience'].apply(lambda x: conv(x))
```

```python
X.head()
```

```python
X.info()
```

```python

Y = df.iloc[:,-1]
print(Y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X, Y)
```

```python
y_pred = lr.predict(X)
y_pred
```

```python
Y
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred, Y)
```

```python
lr.predict([[3,9,7]])
```
